```python
import cv2 as cv
import os
import mediapipe as mp
import tensorflow as tf
import time
from matplotlib import pyplot as plt
```

```python
mp_holistic=mp.solutions.holistic
mp_drawing=mp.solutions.drawing_utils
```

```python
def detect(image,model):
    image=cv.cvtColor(image,cv.COLOR_BGR2RGB)
    result=model.process(image)
    image=cv.cvtColor(image,cv.COLOR_RGB2BGR)
    return image,result
```

```python
def draw_landmark(image,results,dot_size=0):
    mp_drawing.draw_landmarks(image,results.face_landmarks,mp_holistic.FACEMESH_CONTOURS,
                              mp_drawing.DrawingSpec(color=(0,0,255),thickness=1,circle_radius=1),
                              mp_drawing.DrawingSpec(color=(0,255,0),thickness=1,circle_radius=1))
    mp_drawing.draw_landmarks(image,results.pose_landmarks,mp_holistic.POSE_CONNECTIONS,
                              mp_drawing.DrawingSpec(color=(0,0,255),thickness=1,circle_radius=1),
                              mp_drawing.DrawingSpec(color=(0,255,255),thickness=1,circle_radius=1))
    mp_drawing.draw_landmarks(image,results.right_hand_landmarks,mp_holistic.HAND_CONNECTIONS,
                              mp_drawing.DrawingSpec(color=(0,0,255),thickness=1,circle_radius=1),
                              mp_drawing.DrawingSpec(color=(255,255,0),thickness=1,circle_radius=1))
    mp_drawing.draw_landmarks(image,results.left_hand_landmarks,mp_holistic.HAND_CONNECTIONS,
                              mp_drawing.DrawingSpec(color=(0,0,255),thickness=1,circle_radius=1),
                              mp_drawing.DrawingSpec(color=(255,255,0),thickness=1,circle_radius=1))
```

```python
cap=cv.VideoCapture(0)
with mp_holistic.Holistic(min_detection_confidence=0.5,min_tracking_confidence=0.5) as holistic:
    while cap.isOpened():
        ret , frame =cap.read()
        image,results=detect(frame,holistic)
        draw_landmark(image,results)
        cv.imshow("AsITis",image)
        if(cv.waitKey(10) & 0xFF == 27):  # 27 is the ASCII code for the Escape key
            break

        
cap.release()
cv.destroyAllWindows()
```
